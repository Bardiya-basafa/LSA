# Question 1: TF and IDF – Definition, Formulas, and Why They Should Not Be Used Independently

## Term Frequency (TF)

Term Frequency (TF) measures how frequently a word appears in a document. The intuition is that words appearing many times in a document are likely to be important for that document.

### Mathematical Formula

Let:

- $t$ = a term (word)
- $d$ = a document
- $f(t,d)$ = number of occurrences of term $t$ in document $d$

The simplest form of TF is

$$
TF(t,d)=f(t,d)
$$

A normalized version is

$$
TF(t,d)=\frac{f(t,d)}{\sum_{w\in d}f(w,d)}
$$

where the denominator is the total number of words in the document.

### Interpretation

- High TF indicates that a word appears frequently in a document.
- Low TF indicates that a word appears rarely.

### Example

Document:

> "Machine learning improves machine intelligence."

Word counts:

- Machine = 2
- Learning = 1
- Improves = 1
- Intelligence = 1

Total words = 5

Therefore,

$$
TF(\text{Machine})=\frac{2}{5}=0.4
$$

---

## Inverse Document Frequency (IDF)

Inverse Document Frequency (IDF) measures how unique or rare a word is across an entire collection of documents (called a corpus).

Very common words such as **the**, **is**, and **of** appear in almost every document and therefore receive low importance.

### Mathematical Formula

Let

- $N$ = total number of documents
- $df(t)$ = number of documents containing term $t$

Then

$$
IDF(t)=\log\left(\frac{N}{df(t)}\right)
$$

A commonly used smoothed version is

$$
IDF(t)=\log\left(\frac{N+1}{df(t)+1}\right)+1
$$

### Interpretation

- Rare words have high IDF.
- Common words have low IDF.

### Example

Suppose there are 1000 documents.

The word **computer** appears in 40 documents.

$$
IDF=\log\left(\frac{1000}{40}\right)=3.22
$$

The word **the** appears in 998 documents.

$$
IDF=\log\left(\frac{1000}{998}\right)\approx0
$$

Therefore, **computer** is considered much more informative than **the**.

---

## TF-IDF

TF and IDF are usually combined into a single weighting scheme called TF-IDF.

The formula is

$$
TFIDF(t,d)=TF(t,d)\times IDF(t)
$$

A word receives a high TF-IDF score only if it appears frequently in a document but rarely across the entire corpus.

---

## Why TF and IDF Should Not Be Studied Independently

### Limitations of TF Alone

TF only measures frequency within one document.

A common word like **the** may appear many times and receive a high TF score, even though it carries little semantic meaning.

Therefore, TF alone may incorrectly identify common words as important.

### Limitations of IDF Alone

IDF only measures rarity across documents.

A very rare word may receive a high IDF score even if it appears only once due to a spelling mistake or random identifier, making it unimportant for describing the document.

### Why TF-IDF Is Better

TF captures local importance within a document, while IDF captures global uniqueness across the corpus.

By multiplying the two values, TF-IDF identifies words that are both frequent in a document and uncommon across all documents.

Because of this balance, TF-IDF is widely used in:

- Search engines
- Information retrieval
- Text mining
- Natural language processing (NLP)
- Document classification

# Question 2: The Elbow Point Approach for Selecting Singular Values in Truncated SVD

## Background

Singular Value Decomposition (SVD) factorizes a matrix as

$$
A=U\Sigma V^T
$$

where

- $U$ contains the left singular vectors.
- $\Sigma$ contains the singular values.
- $V$ contains the right singular vectors.

The singular values are arranged in descending order:

$$
\sigma_1\ge\sigma_2\ge\cdots\ge\sigma_r
$$

The larger singular values capture most of the important information in the data, while the smaller singular values often represent noise.

Truncated SVD keeps only the first $k$ singular values and their corresponding vectors.

---

## The Challenge

The main challenge is selecting the appropriate value of $k$.

- If $k$ is too small, important information is lost.
- If $k$ is too large, unnecessary noise is retained.

---

## The Elbow Point Method

One common approach is the **Elbow Point Method**.

The singular values are plotted against their indices.

Initially, the singular values decrease rapidly.

After a certain point, the rate of decrease becomes much slower.

This turning point is called the **Elbow Point**.

The value of $k$ is selected at this point because:

- Most of the important information has already been captured.
- The remaining singular values mainly represent noise.

Mathematically,

$$
k=\text{Index of the Elbow Point}
$$

---

## Advantages

- Simple and intuitive.
- Easy to visualize.
- Does not require complicated parameter tuning.
- Works well for many practical datasets.

---

## Limitations

- Some datasets do not have a clear elbow.
- Different users may choose slightly different elbow points.
- It is a heuristic rather than an exact mathematical rule.

---

## Applications

The Elbow Point Method is commonly used in:

- Latent Semantic Analysis (LSA)
- Principal Component Analysis (PCA)
- Image compression
- Recommendation systems
- Dimensionality reduction

# Question 3: Reconstruction Error in Truncated SVD

Suppose a matrix is decomposed using Singular Value Decomposition:

$$
A=U\Sigma V^T
$$

After keeping only the first $k$ singular values, the approximation becomes

$$
A_k=U_k\Sigma_kV_k^T
$$

The difference between the original matrix and its approximation is called the **reconstruction error**.

---

## Frobenius Norm

The most common measure of reconstruction error is the Frobenius norm:

$$
E=\|A-A_k\|_F
$$

where

$$
\|X\|_F=
\sqrt{\sum_i\sum_jx_{ij}^2}
$$

---

## Squared Reconstruction Error

It is often written as

$$
E^2=\|A-A_k\|_F^2
$$

An important property of SVD states that

$$
\|A-A_k\|_F^2=
\sum_{i=k+1}^{r}\sigma_i^2
$$

This means that the reconstruction error equals the sum of the squares of the discarded singular values.

---

## Interpretation

- A small reconstruction error means that the reduced matrix preserves most of the original information.
- A large reconstruction error indicates that too much information has been lost.

---

## Explained Variance

The proportion of preserved information is often measured by

$$
\frac{\sum_{i=1}^{k}\sigma_i^2}
{\sum_{i=1}^{r}\sigma_i^2}
$$

For example, if the explained variance is 95%, then only 5% of the total information has been lost during dimensionality reduction.

# Question 4: Cosine Similarity and Euclidean Distance

Suppose two numerical vectors are

$$
A=(a_1,a_2,\ldots,a_n)
$$

and

$$
B=(b_1,b_2,\ldots,b_n)
$$

---

## Cosine Similarity

Cosine Similarity measures the angle between two vectors rather than their physical distance.

Its mathematical formula is

$$
\cos(\theta)=
\frac{A\cdot B}
{\|A\|\|B\|}
$$

where

$$
A\cdot B=\sum_i a_ib_i
$$

and

$$
\|A\|=
\sqrt{\sum_i a_i^2}
$$

### Interpretation

- A value of **1** means the vectors point in exactly the same direction.
- A value of **0** means the vectors are perpendicular (unrelated).
- A value of **−1** means the vectors point in opposite directions.

For TF-IDF vectors, whose values are non-negative, the similarity usually ranges from **0 to 1**.

---

## Euclidean Distance

Euclidean Distance measures the straight-line distance between two vectors.

Its mathematical formula is

$$
d(A,B)=
\sqrt{\sum_i(a_i-b_i)^2}
$$

### Interpretation

- A distance of **0** means the vectors are identical.
- Larger distances indicate that the vectors are less similar.

Unlike Cosine Similarity, Euclidean Distance has no fixed maximum value because it depends on the scale of the data.

---

## Comparison

| Feature | Cosine Similarity | Euclidean Distance |
|----------|------------------|--------------------|
| Measures | Angle between vectors | Straight-line distance |
| Depends on magnitude | No | Yes |
| Best for | Text mining and document similarity | Numerical coordinates and clustering |
| Minimum | -1 (or 0 for non-negative vectors) | 0 |
| Maximum | 1 | No fixed upper bound |

# Question 5: Standardization of Numerical Vectors and Its Importance Before Matrix Decomposition

## What is Standardization?

Standardization transforms each numerical feature so that it has

- Mean = 0
- Standard deviation = 1

This process removes differences in measurement scales among features.

---

## Mathematical Formula

For a feature value $x$,

$$
z=\frac{x-\mu}{\sigma}
$$

where

- $\mu$ is the mean of the feature.
- $\sigma$ is the standard deviation.
- $z$ is the standardized value (z-score).

---

## Example

Suppose two features are

- Age: values between 18 and 70
- Income: values between 10,000 and 1,000,000

Without standardization, the Income feature dominates because its numerical values are much larger.

After standardization, both features have equal influence during analysis.

---

## Why Standardization Is Important Before SVD

Singular Value Decomposition is based on the variance of the data.

If one feature has much larger values than the others, it contributes disproportionately to the singular values and dominates the decomposition.

As a result, important information contained in smaller-scale features may be ignored.

Standardization ensures that all features contribute fairly to the decomposition.

---

## Advantages

- Prevents large-scale variables from dominating.
- Improves numerical stability.
- Produces more meaningful principal components.
- Improves clustering and dimensionality reduction.
- Enhances the performance of many machine learning algorithms.

---

## When Standardization May Not Be Necessary

Standardization may not be required when

- All features already have similar scales.
- The original scale has physical meaning that should be preserved.
- The data are already normalized, such as TF-IDF vectors.

# Question 6: Randomized SVD – Algorithm, Working Principle, and Applications

## Introduction

Traditional Singular Value Decomposition (SVD) computes an exact decomposition of a matrix.

For very large matrices, exact SVD becomes computationally expensive.

Randomized SVD is an approximate algorithm that efficiently computes only the dominant singular values and vectors.

---

## Main Idea

Instead of directly decomposing the original matrix, Randomized SVD first projects the data into a smaller random subspace that preserves most of the important information.

The SVD is then computed on this much smaller matrix.

---

## Pseudocode

```text
Input:
    Matrix A (m × n)
    Target rank k
    Oversampling parameter p

1. Generate a random Gaussian matrix Ω of size (n × (k+p))

2. Compute
       Y = AΩ

3. Compute the QR decomposition
       Y = QR

4. Project the matrix
       B = QᵀA

5. Compute the exact SVD
       B = ÛΣVᵀ

6. Recover the left singular vectors
       U = QÛ

Output:
       U, Σ, V
```

---

## How the Algorithm Works

1. Generate random vectors.
2. Multiply the original matrix by these random vectors.
3. Construct an orthonormal basis using QR decomposition.
4. Project the original matrix into the lower-dimensional subspace.
5. Compute the exact SVD of the smaller matrix.
6. Map the results back to the original space.

Because the smaller matrix is much easier to process, the algorithm is significantly faster than classical SVD.

---

## Time Complexity

Classical SVD requires approximately

$$
O(mn\min(m,n))
$$

operations.

Randomized SVD reduces the complexity to approximately

$$
O(mnk)
$$

where

$$
k\ll\min(m,n)
$$

Therefore, Randomized SVD is much faster when only a low-rank approximation is needed.

---

## Advantages

- Much faster than exact SVD for large datasets.
- Uses less memory.
- Produces high-quality low-rank approximations.
- Works well with sparse matrices.
- Scales efficiently to very large datasets.

---

## Limitations

- Produces an approximation instead of an exact decomposition.
- Accuracy depends on the chosen target rank.
- Some datasets require additional power iterations for higher accuracy.

---

## Applications

Randomized SVD is widely used in

- Latent Semantic Analysis (LSA)
- Principal Component Analysis (PCA)
- Recommendation systems
- Image compression
- Natural Language Processing (NLP)
- Computer vision
- Bioinformatics
- Large-scale scientific computing

It has become one of the standard methods for efficient dimensionality reduction because it provides an excellent balance between computational speed and approximation accuracy.

# Question 7 - Reading and Preprocessing the Dataset

## Objective

Before performing text mining and Latent Semantic Indexing (LSI), the dataset must be preprocessed to obtain a clean and consistent representation of the text.

Although the assignment only requires converting the text to lowercase and removing punctuation, additional preprocessing steps are commonly used in Natural Language Processing (NLP) to improve the quality of the extracted features.

The preprocessing pipeline implemented in this project includes:

1. Reading the dataset.
2. Converting all characters to lowercase.
3. Removing HTML tags.
4. Removing URLs.
5. Removing email addresses.
6. Removing numbers.
7. Removing punctuation.
8. Removing extra whitespace.
9. Tokenizing the text into words.
10. Removing English stop words.
11. Lemmatizing each word.
12. Removing very short words (less than three characters).

The resulting cleaned text is stored in a new column named **CleanText** and will be used throughout the remainder of the project.

````python
# ============================================================
# Question 7
# Reading and preprocessing the dataset
# ============================================================

import pandas as pd
import numpy as np
import re
import string

import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# ------------------------------------------------------------
# Download NLTK resources (only required once)
# ------------------------------------------------------------

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")

# ------------------------------------------------------------
# Read Dataset
# ------------------------------------------------------------

df = pd.read_csv("dataset.csv")

print("Dataset Shape:", df.shape)

display(df.head())

# ------------------------------------------------------------
# Initialize NLP tools
# ------------------------------------------------------------

stop_words = set(stopwords.words("english"))

custom_stopwords = {
    "said",
    "would",
    "could",
    "also",
    "one",
    "two",
    "may",
    "might",
    "mr",
    "mrs",
    "ms",
    "told",
    "say",
    "says",
    "made",
    "make",
    "get",
    "got",
    "take",
    "way",
    "last",
    "first",
    "new",
}

stop_words.update(custom_stopwords)

lemmatizer = WordNetLemmatizer()

# ------------------------------------------------------------
# Text preprocessing function
# ------------------------------------------------------------


def preprocess_text(text):

    text = str(text)

    # Lowercase
    text = text.lower()

    # Remove HTML tags
    text = re.sub(r"<.*?>", " ", text)

    # Remove URLs
    text = re.sub(r"http\S+|www\S+", " ", text)

    # Remove email addresses
    text = re.sub(r"\S+@\S+", " ", text)

    # Remove numbers
    text = re.sub(r"\d+", " ", text)

    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))

    # Remove extra whitespace
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    tokens = word_tokenize(text)

    cleaned_tokens = []

    for token in tokens:

        # Remove stop words
        if token in stop_words:
            continue

        # Remove short words
        if len(token) < 3:
            continue

        # Lemmatization
        token = lemmatizer.lemmatize(token)

        cleaned_tokens.append(token)

    return " ".join(cleaned_tokens)


# ------------------------------------------------------------
# Apply preprocessing
# ------------------------------------------------------------

df["CleanText"] = df["Text"].apply(preprocess_text)

print("\nPreprocessing completed successfully.\n")

display(df[["Text", "CleanText"]].head())

# ------------------------------------------------------------
# Save cleaned dataset
# ------------------------------------------------------------

df.to_csv("dataset_clean.csv", index=False)

print("Clean dataset saved as dataset_clean.csv")
````

## Explanation

The preprocessing stage transforms every document into a cleaner and more consistent representation.

### Lowercase Conversion

Words such as

- Computer
- COMPUTER
- computer

are converted into a single representation (**computer**), reducing the vocabulary size.

### Removing HTML, URLs, and Email Addresses

These elements usually do not contribute to the semantic meaning of a document and therefore are removed.

### Removing Numbers

Most numerical values (such as years, IDs, and prices) do not improve topic modeling and are removed.

### Removing Punctuation

Punctuation symbols do not represent meaningful terms in a Bag-of-Words model.

### Tokenization

Each document is divided into individual words (tokens).

### Stop-word Removal

Common English words such as

- the
- is
- of
- and
- to

appear in almost every document and therefore provide very little information about the document's topic.

Removing them significantly improves the quality of text analysis.

### Lemmatization

Lemmatization converts different grammatical forms of a word into its base form.

Examples:

- cars → car
- studies → study
- running → running (or run with POS tagging)

This reduces the vocabulary size while preserving the semantic meaning of the words.

The cleaned documents are stored in the **CleanText** column and will be used throughout the remaining questions.

# Question 10 - Constructing the Bag-of-Words Matrix

## Objective

The file **words.csv** contains a predefined vocabulary sorted alphabetically.

The goal is to construct a **Bag-of-Words (BoW)** representation of the corpus.

For every document, the number of occurrences of each vocabulary word is counted.

The resulting document-term matrix has:

- One row for each document.
- One column for each word in **words.csv**.

Each element of the matrix represents the frequency of a vocabulary word in a document.

````python
# ============================================================
# Question 10
# Bag of Words
# ============================================================

import pandas as pd
import numpy as np

# ------------------------------------------------------------
# Read Vocabulary
# ------------------------------------------------------------

vocabulary = pd.read_csv("words.csv")

# First column contains the vocabulary
vocabulary = vocabulary.iloc[:,0].tolist()

print("Vocabulary Size:", len(vocabulary))

# ------------------------------------------------------------
# Create word -> column index dictionary
# ------------------------------------------------------------

word_to_index = {
    word:index
    for index, word in enumerate(vocabulary)
}

# ------------------------------------------------------------
# Create Bag-of-Words Matrix
# ------------------------------------------------------------

num_documents = len(df)

num_words = len(vocabulary)

bow_matrix = np.zeros((num_documents, num_words), dtype=np.int32)

for doc_index, document in enumerate(df["CleanText"]):

    words = document.split()

    for word in words:

        if word in word_to_index:

            bow_matrix[
                doc_index,
                word_to_index[word]
            ] += 1

# ------------------------------------------------------------
# Convert to DataFrame
# ------------------------------------------------------------

bow_df = pd.DataFrame(
    bow_matrix,
    columns=vocabulary
)

display(bow_df.head())

print()

print("Bag-of-Words Matrix Shape:")

print(bow_df.shape)
````

## Explanation

The Bag-of-Words (BoW) model represents each document as a numerical vector based on the frequency of predefined vocabulary words.

The vocabulary is obtained from **words.csv**, which contains **52 words** sorted alphabetically. Each column of the Bag-of-Words matrix corresponds to one vocabulary word, while each row represents a news article.

For every document, the occurrences of each vocabulary word are counted. If a word does not appear in a document, its corresponding value is **0**; otherwise, the value equals the number of occurrences of that word.

For example, in the first document:

- **chief** appears once.
- **tax** appears eight times.
- **time** appears twice.
- Most of the remaining vocabulary words do not appear, so their values are zero.

This representation ignores the order of words and only preserves their frequencies, which is why it is called the **Bag-of-Words** model.

The dimensions of the resulting document-term matrix are:

- **Number of documents:** 2225
- **Vocabulary size:** 52

Therefore, the Bag-of-Words matrix has dimensions

$$
2225 \times 52
$$

where:

- Each **row** corresponds to one news article.
- Each **column** corresponds to one vocabulary word.
- Each **matrix element** represents the frequency of a vocabulary word in a document.

This document-term matrix serves as the input for the following stages of the project, including standardization, Singular Value Decomposition (SVD), Truncated SVD, and Latent Semantic Indexing (LSI).

# Question 12 - Standardization and Singular Value Decomposition (SVD)

## Objective

The Bag-of-Words features have different numerical ranges because some words occur much more frequently than others.

Before applying Singular Value Decomposition, the features are standardized so that each feature has:

- Mean = 0
- Standard Deviation = 1

After standardization, full Singular Value Decomposition (SVD) is performed.

The dimensions of the resulting matrices are then reported.

````python
# ============================================================
# Question 12
# Standardization + Full SVD
# ============================================================

from sklearn.preprocessing import StandardScaler
import numpy as np

# ------------------------------------------------------------
# Standardization
# ------------------------------------------------------------

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

print("Standardization completed.")

# ------------------------------------------------------------
# Full Singular Value Decomposition
# ------------------------------------------------------------

U, S, VT = np.linalg.svd(
    X_train_std,
    full_matrices=False
)

Sigma = np.diag(S)

# ------------------------------------------------------------
# Report Matrix Dimensions
# ------------------------------------------------------------

print()

print("=" * 40)
print("SVD Matrix Dimensions")
print("=" * 40)

print(f"U      : {U.shape}")

print(f"Sigma  : {Sigma.shape}")

print(f"V^T    : {VT.shape}")
````

## Explanation

Before applying Singular Value Decomposition, the Bag-of-Words matrix is standardized using the z-score transformation.

Standardization is necessary because some words occur much more frequently than others. Without standardization, these high-frequency words would dominate the decomposition and reduce the influence of less frequent but potentially informative words.

After standardization, Singular Value Decomposition factorizes the matrix into three matrices:

$$
A = U \Sigma V^T
$$

where

- **U** contains the left singular vectors (document space).
- **Σ** is a diagonal matrix containing the singular values.
- **Vᵀ** contains the right singular vectors (word space).

Since the training matrix contains **2000 documents** and **52 vocabulary words**, its dimensions are

$$
2000 \times 52
$$

Because

$$
2000 > 52,
$$

the dimensions of the resulting matrices are

| Matrix | Dimensions |
|---------|------------|
| $U$ | $2000 \times 52$ |
| $\Sigma$ | $52 \times 52$ |
| $V^T$ | $52 \times 52$ |

These matrices contain the complete latent semantic representation of the training corpus and are used in the following question to perform dimensionality reduction using Truncated SVD.

# Question 12-2 - Selecting the Rank Using the Elbow Method and Applying Truncated SVD

## Objective

The complete SVD contains all singular values.

However, many of the smaller singular values mainly represent noise.

To obtain a lower-dimensional latent semantic space, an appropriate number of singular values must be selected.

This is achieved using the **Elbow Method**.

````python
# ============================================================
# Question 13
# Elbow Method + Truncated SVD
# ============================================================

import matplotlib.pyplot as plt

# ------------------------------------------------------------
# Plot Singular Values
# ------------------------------------------------------------

plt.figure(figsize=(10,6))

plt.plot(
    range(1, len(S)+1),
    S,
    marker='o'
)

plt.xlabel("Component Number")

plt.ylabel("Singular Value")

plt.title("Elbow Method for Selecting k")

plt.grid(True)

plt.show()

# ------------------------------------------------------------
# Choose the reduced rank based on the elbow plot
# ------------------------------------------------------------

k = 18

print(f"Selected Rank (k): {k}")

# ------------------------------------------------------------
# Construct Truncated SVD
# ------------------------------------------------------------

U_k = U[:, :k]

Sigma_k = Sigma[:k, :k]

VT_k = VT[:k, :]

# ------------------------------------------------------------
# Reconstruct the matrix
# ------------------------------------------------------------

X_reconstructed = U_k @ Sigma_k @ VT_k

# ------------------------------------------------------------
# Reconstruction Error (Frobenius Norm)
# ------------------------------------------------------------

reconstruction_error = np.linalg.norm(
    X_train_std - X_reconstructed,
    ord="fro"
)

print(f"Reconstruction Error: {reconstruction_error:.4f}")
````

## Interpretation

The singular values are plotted in descending order to determine an appropriate reduced rank for Truncated SVD using the Elbow Method.

From the graph, the first few singular values decrease very rapidly, indicating that the corresponding components capture most of the important semantic information contained in the document-term matrix.

After approximately the **18th component**, the curve becomes much flatter and the rate of decrease is significantly reduced. This point represents the **elbow**, where additional components contribute relatively little new information and mostly capture noise or minor variations in the data.

Therefore, the reduced rank is selected as

$$
k = 18.
$$

Using the first 18 singular values provides a good compromise between dimensionality reduction and information preservation.

The reduced approximation of the standardized Bag-of-Words matrix is computed as

$$
A_k = U_k \Sigma_k V_k^T,
$$

where only the first **18** singular values and their corresponding singular vectors are retained.

The quality of this approximation is evaluated using the reconstruction error, computed with the Frobenius norm:

$$
\|A-A_k\|_F.
$$

A smaller reconstruction error indicates that the reduced latent semantic representation preserves more information from the original matrix while using far fewer dimensions.

# Question 13 - Implementation of Randomized SVD

## Objective

When the input matrix is very large, computing the full Singular Value Decomposition (SVD) can be computationally expensive in terms of both execution time and memory usage.

Randomized Singular Value Decomposition (Randomized SVD) is an efficient approximation algorithm that computes only the dominant singular values and singular vectors of a matrix.

Instead of decomposing the original matrix directly, the algorithm first projects the matrix into a lower-dimensional random subspace that approximately preserves its most important information.

After obtaining this smaller representation, the classical SVD is performed on the reduced matrix. Finally, the singular vectors of the original matrix are reconstructed from the reduced decomposition.

Compared to the classical SVD, Randomized SVD has significantly lower computational complexity for large sparse matrices and is widely used in:

- Latent Semantic Indexing (LSI)
- Principal Component Analysis (PCA)
- Recommender Systems
- Information Retrieval
- Machine Learning
- Large-scale Data Mining

## Randomized SVD Algorithm (Pseudo Code)

Given a matrix

$$
A \in \mathbb{R}^{m \times n}
$$

and a target rank

$$
k,
$$

the algorithm proceeds as follows:

```

Input:
    Matrix A
    Target rank k
    Oversampling parameter p

1. Generate a random Gaussian matrix Ω
       Ω ∈ R^(n × (k+p))

2. Compute

       Y = AΩ

3. Compute the QR decomposition

       Y = QR

4. Project A into the low-dimensional space

       B = QᵀA

5. Compute the classical SVD

       B = ÛΣVᵀ

6. Recover the left singular vectors

       U = QÛ

7. Keep the first k components.

Output:
    U_k
    Σ_k
    V_kᵀ

```

This algorithm approximates the dominant singular vectors while requiring much less computation than the full SVD for large matrices.

````python
# ============================================================
# Question 13
# Randomized SVD Implementation
# ============================================================

import numpy as np

def randomized_svd(A, k, oversampling=10, random_state=42):
    """
    Randomized Singular Value Decomposition

    Parameters
    ----------
    A : ndarray
        Input matrix

    k : int
        Target rank

    oversampling : int
        Additional random vectors

    random_state : int
        Random seed

    Returns
    -------
    U_k
    Sigma_k
    VT_k
    """

    np.random.seed(random_state)

    # --------------------------------------------------------
    # Step 1
    # Generate random Gaussian matrix
    # --------------------------------------------------------

    Omega = np.random.randn(
        A.shape[1],
        k + oversampling
    )

    # --------------------------------------------------------
    # Step 2
    # Random Projection
    # --------------------------------------------------------

    Y = A @ Omega

    # --------------------------------------------------------
    # Step 3
    # QR Decomposition
    # --------------------------------------------------------

    Q, _ = np.linalg.qr(Y)

    # --------------------------------------------------------
    # Step 4
    # Project A
    # --------------------------------------------------------

    B = Q.T @ A

    # --------------------------------------------------------
    # Step 5
    # Classical SVD
    # --------------------------------------------------------

    U_hat, S, VT = np.linalg.svd(
        B,
        full_matrices=False
    )

    # --------------------------------------------------------
    # Step 6
    # Recover left singular vectors
    # --------------------------------------------------------

    U = Q @ U_hat

    # --------------------------------------------------------
    # Step 7
    # Keep only first k components
    # --------------------------------------------------------

    U_k = U[:, :k]

    Sigma_k = np.diag(S[:k])

    VT_k = VT[:k, :]

    return U_k, Sigma_k, VT_k
````

````python
# ------------------------------------------------------------
# Run Randomized SVD
# ------------------------------------------------------------

k = 18

U_rand, Sigma_rand, VT_rand = randomized_svd(
    X_train_std,
    k
)

print("="*40)
print("Randomized SVD Results")
print("="*40)

print("U Shape      :", U_rand.shape)
print("Sigma Shape  :", Sigma_rand.shape)
print("VT Shape     :", VT_rand.shape)
````

## Explanation

The implemented Randomized SVD algorithm follows the standard procedure proposed by Halko et al.

Instead of directly decomposing the original document-term matrix, the algorithm first projects the matrix into a lower-dimensional random subspace using a Gaussian random matrix.

A QR decomposition is then performed to obtain an orthonormal basis that approximates the dominant column space of the original matrix.

The original matrix is projected onto this basis, producing a much smaller matrix. A conventional Singular Value Decomposition is then applied to this reduced matrix.

Finally, the left singular vectors of the original matrix are reconstructed by multiplying the orthonormal basis with the left singular vectors obtained from the reduced decomposition.

Only the first **k = 18** singular values and their corresponding singular vectors are retained.

The resulting matrices have the following dimensions:

- **U:** (2000 × 18)
- **Σ:** (18 × 18)
- **Vᵀ:** (18 × 52)

These matrices provide a low-dimensional approximation of the standardized Bag-of-Words matrix and are used in the next question to compare Randomized SVD with Truncated SVD.

The Truncated SVD achieved a reconstruction error of **205.452685**, whereas the Randomized SVD produced a slightly larger reconstruction error of **225.098708**.

This result is expected because Truncated SVD is computed from the exact Singular Value Decomposition and therefore provides the optimal rank-18 approximation of the original matrix according to the Eckart–Young theorem.

Randomized SVD approximates the dominant singular vectors using random projections. Consequently, a small loss in reconstruction accuracy is expected.

### Execution Time

For this dataset, Truncated SVD completed in approximately **0.0021 seconds**, while Randomized SVD required approximately **0.0076 seconds**.

Although Randomized SVD was slower in this experiment, this is mainly because the dataset is relatively small (2000 documents and 52 features). ### Which Method is Better?

For relatively small datasets such as this one, **Truncated SVD is the preferable method** because:

- it provides the optimal low-rank approximation,
- it achieves a smaller reconstruction error,
- and it is faster for matrices of this size.

However, if the dataset contained millions of documents, such as the entire collection of text available on the Internet, **Randomized SVD would be the preferred approach**. In very large and sparse matrices, computing the exact SVD becomes computationally expensive in terms of both execution time and memory usage. Randomized SVD significantly reduces these computational requirements while still producing an approximation that is sufficiently accurate for most Natural Language Processing and Information Retrieval applications.

Therefore, the choice of decomposition method depends on the size of the dataset:

- **Small or medium-sized datasets:** Truncated SVD is preferred due to its higher accuracy and lower computational cost.
- **Very large datasets:** Randomized SVD is recommended because of its superior scalability and memory efficiency, despite introducing a small approximation error.

## Objective

After applying Truncated SVD, every vocabulary word is represented as a vector in the latent semantic space.

In this question, the similarity of the following word pairs is measured using two different metrics:

- (director, film)
- (win, won)
- (play, game)
- (play, law)
- (government, music)

The following similarity measures are used:

- Cosine Similarity
- Euclidean Distance

The obtained values are then interpreted to evaluate the semantic relationships between the selected words.

````python
# ============================================================
# Question 16
# Word Similarity in the Latent Space
# ============================================================

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

# ------------------------------------------------------------
# Word vectors in latent space
# ------------------------------------------------------------
# Each row corresponds to a word represented in the latent space.
# VT_k has shape (k × vocabulary_size), so transpose it.

word_vectors = VT_k.T

# ------------------------------------------------------------
# Vocabulary dictionary
# ------------------------------------------------------------

word_index = {
    word: index
    for index, word in enumerate(vocabulary)
}

# ------------------------------------------------------------
# Word pairs
# ------------------------------------------------------------

word_pairs = [

    ("mobile", "technology"),
    ("director", "film"),
    ("win", "won"),
    ("play", "game"),
    ("play", "law"),
    ("government", "music")

]

# ------------------------------------------------------------
# Compute similarities
# ------------------------------------------------------------

results = []

for word1, word2 in word_pairs:

    if word1 not in word_index or word2 not in word_index:

        results.append([
            word1,
            word2,
            "Not Found",
            "Not Found"
        ])

        continue

    v1 = word_vectors[word_index[word1]]
    v2 = word_vectors[word_index[word2]]

    cosine = cosine_similarity(
        [v1],
        [v2]
    )[0][0]

    euclidean = np.linalg.norm(v1 - v2)

    results.append([
        word1,
        word2,
        cosine,
        euclidean
    ])

similarity_df = pd.DataFrame(
    results,
    columns=[
        "Word 1",
        "Word 2",
        "Cosine Similarity",
        "Euclidean Distance"
    ]
)

display(similarity_df)
````

# Question 18 - Searching in the Latent Semantic Space

## Objective

Suppose we want to search for the word **technology** in a text database. ## Explanation

In the traditional **Bag-of-Words (BoW)** model, every word is treated as an independent feature. A document is considered relevant to a query only if it contains the exact query term or if another explicit matching mechanism is used.

For example, if a user searches for the word **technology**, a document containing only the words **digital** and **mobile** may receive a very low similarity score or may not be retrieved at all because the exact word **technology** does not appear in that document.

Latent Semantic Indexing (LSI) addresses this limitation by projecting both documents and words into a lower-dimensional **latent semantic space** using Singular Value Decomposition (SVD).

In this latent space, words that frequently occur in similar contexts are positioned close to one another. Consequently, words such as:

- technology
- digital
- mobile
- computer
- phone

are likely to have similar vector representations because they commonly appear in technology-related documents.

Therefore, even if a document does not explicitly contain the word **technology**, it can still be retrieved because its latent vector is close to the latent vector of the query.

For example:

**Query:**

```
technology
```

**Document A:**

```
mobile phone digital service computer
```

Although the word **technology** never appears in Document A, the document is still considered relevant because the words **mobile**, **digital**, **computer**, and **phone** are strongly associated with the same underlying semantic concept.

This is one of the main advantages of LSI over the traditional Bag-of-Words representation.

## Computational Advantages

Searching in the latent semantic space is also computationally more efficient.

In the Bag-of-Words representation, each document is represented by a vector whose dimension equals the size of the vocabulary.

For this project:

- Vocabulary size = **52**
- Document vector dimension = **52**

After applying Truncated SVD:

- Latent dimension = **18**

Thus, every document is represented using only **18 numerical values** instead of **52**.

When the vocabulary contains tens or hundreds of thousands of words—as in real-world search engines—the reduction becomes much more significant.

Searching in a lower-dimensional space provides several advantages:

- Lower memory consumption.
- Faster similarity computations.
- Reduced storage requirements.
- Better scalability for large document collections.
- Reduced influence of noisy or redundant features.

Although Truncated SVD requires an initial computational cost to decompose the matrix, the decomposition is performed only once. Afterwards, all search queries are executed in the reduced latent space, making information retrieval significantly faster.

## Conclusion

Compared with the Bag-of-Words model, Latent Semantic Indexing provides both semantic and computational advantages.

Semantically, it retrieves documents that contain conceptually related words even when the exact query term is absent.

Computationally, it represents documents in a much lower-dimensional space, reducing both storage requirements and the cost of similarity calculations while improving the scalability of large text retrieval systems.

````python
# ============================================================
# Question 19
# Compute Average Latent Vector for Each Class
# ============================================================

import numpy as np
import pandas as pd

# ------------------------------------------------------------
# Document coordinates in latent space
# ------------------------------------------------------------

latent_documents = U_k @ Sigma_k

# ------------------------------------------------------------
# Training labels
# ------------------------------------------------------------

train_labels = y_train.reset_index(drop=True)

# ------------------------------------------------------------
# Compute class centroids
# ------------------------------------------------------------

class_centroids = []

for label in sorted(train_labels.unique()):

    centroid = latent_documents[train_labels == label].mean(axis=0)

    class_centroids.append(centroid)

class_centroids = np.array(class_centroids)

centroid_df = pd.DataFrame(
    class_centroids,
    index=[
        "Politics",
        "Sport",
        "Technology",
        "Entertainment",
        "Business"
    ],
    columns=[
        f"Component {i+1}"
        for i in range(k)
    ]
)

display(centroid_df)
````

### Proposed Classification Method

A simple and effective classification method is the **Nearest Centroid Classifier**.

The procedure is:

1. Preprocess a new document using the same preprocessing pipeline.
2. Convert the document into its Bag-of-Words representation.
3. Standardize the vector using the previously fitted scaler.
4. Project the document into the latent semantic space using the Truncated SVD model.
5. Compute the cosine similarity (or Euclidean distance) between the document vector and each class centroid.
6. Assign the document to the class with the highest cosine similarity (or the smallest Euclidean distance).

This approach does not require interpreting the meaning of individual latent components. Instead, it relies on the overall position of documents in the latent semantic space, making it scalable and suitable for large text collections.
